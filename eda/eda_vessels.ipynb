```python
# Distribution of Numerical Features
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
vessels.hist(bins=30, figsize=(15, 10))
plt.show()

# Box plots
plt.figure(figsize=(15, 10))
sns.boxplot(data=vessels)
plt.xticks(rotation=90)
plt.show()

# Categorical Features Analysis
# Assuming 'vesselType' is a categorical feature
plt.figure(figsize=(10, 6))
sns.countplot(x='vesselType', data=vessels)
plt.xticks(rotation=90)
plt.show()

vessels_corr_matrix = vessels.drop(columns=['shippingLineId', 'vesselId', 'homePort'])

# Correlation Analysis
corr_matrix = vessels_corr_matrix.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()

# Missing Values Analysis
plt.figure(figsize=(15, 10))
sns.heatmap(vessels.isnull(), cbar=False, cmap='viridis')
plt.show()

# Feature Relationships
# Scatter plot example
plt.figure(figsize=(10, 6))
sns.scatterplot(x='length', y='maxSpeed', data=vessels)
plt.show()

# Pair plots - not very useful
# sns.pairplot(vessels)
# plt.show()
```
